# Retail Business Performance & Profitability Analysis
SQL + Python + Visualization Project

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('retail_analysis_dataset.csv')
df['Profit'] = df['Sales'] - df['Cost']
df.head()
```

```
# Correlation
df[['Sales','Cost','Profit','Inventory_Days']].corr()
```

```
# Scatter Plot
plt.figure()
plt.scatter(df['Inventory_Days'], df['Profit'])
plt.title('Inventory Days vs Profit')
plt.xlabel('Inventory Days')
plt.ylabel('Profit')
plt.show()
```
